```python
#install packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import datasets
dfState = pd.read_csv('db/StateNames.csv')
dfNational = pd.read_csv('db/NationalNames.csv')
```

```python
#state dataset
#I only care about Wisconsin if I use this dataset
dfWisc = dfState[dfState['State'] == 'WI']
print(dfWisc.columns)
#Let's start by finding the most popular names all time
top10NameCount = dfWisc[['Name','Count','Year']].sort_values('Count',ascending=False)[:10]
print(top10NameCount.head(5))
top10NameCount.plot(kind='bar', x="Name", y="Count", title='Wisconsin Counts of Unique Baby Names', figsize=(15,10),legend=True,fontsize=12)
```
